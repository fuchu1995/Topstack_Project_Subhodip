```python
df2=pd.get_dummies(df,columns=['parameter'],dtype=int)
df2.head()
```

```python
#Show a distribution for each country the total energy produced from various sources.
def product_trend(country,product):
    d1=df[df['country_name']==country]
    d2=d1[d1['product']==product]
    sns.lineplot(x='date',y='value',data=d2)
    plt.grid()
    plt.show()
```

```python
product_trend('India','Solar')
```

```python
max_export_country = df2.groupby('country_name')['parameter_Total Exports'].sum().idxmax()
print("Country with maximum electricity export business:", max_export_country)
```

```python
df['product'] = pd.to_numeric(df['product'], errors='coerce')
```

```python
df['product'] = df['product'].astype(float)
```

```python
X = df.groupby('parameter')['product'].sum().reset_index()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X[['product']])
kmeans = KMeans(n_clusters=3) # Adjust the number of clusters as needed
kmeans.fit(X_scaled)
X['cluster'] = kmeans.labels_

print("Cluster centers:", scaler.inverse_transform(kmeans.cluster_centers_))
print("Cluster labels:", kmeans.labels_)
```
